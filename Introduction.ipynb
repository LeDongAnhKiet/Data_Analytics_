```python
# Nơi import các thư viện hỗ trợ DA
import numpy as np
```

```python
# Tạo 1 mảng
arr = np.array([1,2,3])
print(arr, '\n')  # In tất cả phần tử trong mảng
print(type(arr), '\n')  # In kiểu dữ liệu
print(arr.shape, '\n')  # In kích thước mảng
```

```python
print(arr[2],'\n') # In phần tử thứ 2 (đếm từ 0)
arr[2] = 5  # Gán giá trị mới vào phần tử thứ 2
print(arr, '\n')
```

```python
arr = np.array([[1,2,3],[4,5,6]]) # Mảng 2 chiều
print(arr, '\n')
print(type(arr), '\n')
print(arr.shape, '\n')
print(arr.ndim, '\n') # Số chiều
```

```python
# Gán giá trị từ 0 đến 50, cách nhau 5 giá trị
arr = np.arange(0, 50, 5)
print(arr, '\n')

# Gán ngẫu nhiên giá trị
arr = np.random.rand(10)
print(arr, '\n')

# Gán chỉ giá trị 0
arr = np.zeros((3,4))
print(arr, '\n')

# Gán chỉ giá trị 1
arr = np.ones((2,3,5))
print(arr, '\n')
```

```python
arr =[1,2,3]
print(np.repeat(arr, 3), '\n')  # Lặp các phần tử 3 lần
print(np.tile(arr, 3), '\n')  # Lặp mảng 3 lần

arr = np.eye(3) # ma trận đơn vị
print(arr, '\n')

arr = np.diag([3,5,7,9]) # ma trận chéo
print(arr, '\n')
```

```python
arr = np.random.rand(5,5)
print(arr,'\n')
print(np.diag(arr),'\n') # Lấy đường chéo của arr thành mảng

print(arr+10,'\n')  # Cộng tất cả phần tử lên 10
print(arr*2,'\n') # Nhân tất cả phần tử cho 2
print(arr/3,'\n') # Chia tất cả phần tử cho 3

print(np.exp(arr),'\n') # Trả về e^x
print(np.log(arr),'\n') # Trả về log^e(x)
print(np.log2(arr),'\n')  # Trả về log^2(x)
print(np.log10(arr),'\n') # Trả về log^10(x)

print(np.sin(arr),'\n') # Trả về sin(x)
print(np.cos(arr),'\n') # Trả về cos(x)
print(np.tan(arr),'\n') # Trả về (x)
```

```python
print(np.sum(arr),'\n') # Tổng các phần tử
print(np.sum(arr, axis=0),'\n') # Tổng các cột
print(np.sum(arr, axis=1),'\n') # Tổng các dòng

print(np.min(arr),'\n') # Số nhỏ nhất
print(np.max(arr),'\n') # Số lớn nhất
```

```python
print(np.mean(arr),'\n')  # Trung bình cộng
print(np.mean(arr, axis=0),'\n')  # Trung bình cộng cột
print(np.median(arr),'\n')  # Số ở giữa mảng
print(np.median(arr, axis=1),'\n')  # Số ở giữa hàng
print(np.std(arr),'\n') # Độ lệch
print(np.var(arr),'\n') # Phương sai

print(np.sort(arr,axis=0),'\n') # Sắp xếp theo cột
print(arr.T,'\n') # Lật cột thành hàng, hàng thành cột
print(arr.flatten(),'\n') # Trả về mảng 1 chiều
```
