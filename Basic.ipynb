# **Tiền xử lý dữ liệu**

```python
# Bỏ các cột không quan trọng
df = car.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style',
               'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
# Đổi tên cột
df = df.rename(columns={'Engine HP': 'Mã lực', 'Engine Cylinders': 'Xi-lanh',
                         'Transmission Type': 'Truyền tải', 'Driven_Wheels': 'Mode',
                         'highway MPG': 'Tiêu thụ ko ngừng',
                         'city mpg': 'Tiêu thụ thường', 'MSRP': 'Giá thành'})
df.head()
```

```python
print(df.shape,'\n') # Số hàng và cột
dup_car = df[df.duplicated()] # Tìm các thông tin trùng lắp
print(dup_car.shape,'\n')

print(df.count(),'\n') # Đếm các hàng không rỗng từng cột
df = df.drop_duplicates() # Xóa các thông tin trùng lắp
print(df.count(), '\n')

print(df.isnull().sum()) # Đếm các thông tin rỗng từng cột
df = df.dropna() # Xóa các hàng rỗng
print(df.count(), '\n')
```

# **Phân tích dữ liệu**

```python
car = df
sns.boxplot(x=car['Giá thành'])
```

```python
sns.boxplot(x=car['Mã lực'])
```

```python
# Biểu đồ số xe mỗi thương hiệu
car.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
# Đặt tiêu đề
plt.title('Số lượng sản xuất')
plt.xlabel('Thương hiệu')
plt.ylabel('Số lượng xe')
```

# **Phân tích mối tương quan**

```python
plt.figure(figsize=(20,10))
c = car.corr() # Mối liên hệ
sns.heatmap(c, cmap='BrBG', annot=True) # Biểu đồ thể hiện độ tương quan
```
